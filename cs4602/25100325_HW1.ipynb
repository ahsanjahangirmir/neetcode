##### **Problem 1:**
Pair with Target Sum (Ungraded, Warm-up)

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:

        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
```

##### **Problem 3:**
Triple sum to zero

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        
        res = set()
        nums.sort()
        n = len(nums)

        for i in range(n):

            if nums[i] > 0:
                break
            
            l = i + 1
            r = n - 1

            while l < r:
                
                if nums[i] + nums[l] + nums[r] == 0:
                    res.add((nums[i], nums[l], nums[r]))
                    l += 1
                    r -= 1
                
                elif nums[i] + nums[l] + nums[r] < 0:
                    l += 1
                
                else:
                    r -= 1

        return [list(tup) for tup in res]
```

##### **Problem 4:** 
Triple sum close to target

```python
class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        
        closest_difference = 1000000
        closest_sum = 0
        n = len(nums)
        nums.sort()

        for i in range(n - 2):
            l = i + 1
            r = n - 1
            while l < r:
                threeSum = nums[i] + nums[l] + nums[r]
                diff = abs(target - threeSum) 

                if diff == 0:
                    return target
                
                if diff < closest_difference:
                    closest_difference = diff
                    closest_sum = threeSum

                if threeSum < target:
                    l += 1
                else:
                    r -= 1
                    

        return closest_sum
```

##### **Problem 5:**
Subarrays with Product less than a target

```python
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:

        n = len(nums)
        product = 1 

        count = 0

        l = r = 0

        for i in range(n):

            product *= nums[r]

            while product >= k and l <= r:
                product //= nums[l]
                l += 1 

            count += r - l + 1
            
            r += 1
        
        return count 
```

##### **Problem 6:** 
Dutch National Flag Problem

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        
        red = 0

        for i in range(len(nums)):
            if nums[i] == 0:
                temp = nums[red]
                nums[red] = nums[i]
                nums[i] = temp 
                red += 1 
        
        white = red 

        for i in range(white, len(nums)):
            if nums[i] == 1:
                temp = nums[white]
                nums[white] = nums[i]
                nums[i] = temp 
                white += 1 

        blue = white

        for i in range(blue, len(nums)):
            if nums[i] == 2:
                temp = nums[blue]
                nums[blue] = nums[i]
                nums[i] = temp 
                blue += 1 
```

##### **Problem 7:** 
Quadruple sum to target

```python
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        
        res = set()
        n = len(nums)
        nums.sort()
        
        for first in range(n):

            second = first + 1 

            while second < n - 2:
                
                l = second + 1
                r = n - 1 
                remainingSum = target - nums[first] - nums[second] 
                
                while l < r:

                    innerSum = nums[l] + nums[r]

                    if innerSum == remainingSum:
                        res.add((nums[first], nums[second], nums[l], nums[r]))
                        r -= 1

                    if innerSum < remainingSum:
                        l += 1
                    
                    if innerSum > remainingSum:
                        r -= 1

                second += 1


        return sorted([list(tup) for tup in res])

```

##### **Problem 8:**
Comparing strings containing backspaces

```python
class Solution:
    def backspaceCompare(self, s: str, t: str) -> bool:

        def helper(stack: list, string: str) -> list:
            
            for ch in string:
                if ch == '#':
                    if len(stack) > 0:
                        stack.pop()
                else:
                    stack.append(ch)

            return stack

        return helper([], s) == helper([], t)
```
