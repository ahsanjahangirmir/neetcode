##### **Problem 1: Palindrome LinkedList**
Write a function that checks if a linked list is a palindrome.


```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        
        if not head.next:
            return True
        
        # get mid of ll 
        
        s = head
        f = head

        while f != None and f.next != None:
            s = s.next
            f = f.next.next
        
        # reverse second half of ll in-place 
        
        prev = None
        curr = s

        while curr: 
            nextNode = curr.next
            curr.next = prev
            prev = curr
            curr = nextNode

        start = head
        mid = prev
        
        #compare halves 
        
        while mid:
            if start.val != mid.val:
                return False
        
            start = start.next
            mid = mid.next

        return True

```

##### **Problem 2: Rearrange a LinkedList**
Rearrange a linked list such that the nodes are ordered in an alternating high-low fashion.


```python
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        if not head.next:
            return True
        
        # get mid of ll 
        
        s = head
        f = head

        while f and f.next:
            s = s.next
            f = f.next.next
        
        # reverse second half of ll in-place 
        
        prev = None
        curr = s

        while curr:
            nextNode = curr.next
            curr.next = prev
            prev = curr
            curr = nextNode

        start = head
        mid = prev
        
        while mid.next:
            temp1 = start.next
            temp2 = mid.next
            start.next = mid
            mid.next = temp1
            start = temp1 
            mid = temp2

        return True
```

##### **Problem 3: Maximum Twin Sum of a Linked List**
Find the maximum twin sum in a linked list.


```python
class Solution:
    def pairSum(self, head: Optional[ListNode]) -> int:

        if not head.next.next:
            return head.val + head.next.val

        # get mid of ll 
        
        s = head
        f = head

        while f != None and f.next != None:
            s = s.next
            f = f.next.next
        
        # reverse second half of ll in-place 
        
        prev = None
        curr = s

        while curr: 
            nextNode = curr.next
            curr.next = prev
            prev = curr
            curr = nextNode

        start = head
        mid = prev
        
        #compare halves 

        maxTwinSum = None
        
        while mid:

            twinSum = start.val + mid.val

            if maxTwinSum is None or twinSum > maxTwinSum:
                maxTwinSum = twinSum
        
            start = start.next
            mid = mid.next
        
        return maxTwinSum
```

##### **Problem 4: Cycle in Circular Array**
Determine if there is a cycle in a circular array of numbers.


```python
class Solution:
    def circularArrayLoop(self, nums: List[int]) -> bool:
        
        def next_index(i):
            return (i + nums[i]) % n
        
        n = len(nums)
        
        for i in range(n):
        
            if nums[i] == 0:
                continue

            slow = i
            fast = next_index(i)
            
            while nums[slow] * nums[fast] > 0 and nums[slow] * nums[next_index(fast)] > 0:
                
                if slow == fast:
                    if slow == next_index(slow):
                        break
                    return True
                
                slow = next_index(slow)
                fast = next_index(next_index(fast))

            slow = i
            direction = nums[i]

            while nums[slow] * direction > 0:
                next_slow = next_index(slow)
                nums[slow] = 0
                slow = next_slow
        
        return False
```
